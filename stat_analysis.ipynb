```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merged_data = pd.read_csv('assets/cardio_train_clean.csv')

# Query the data
smoke_and_drink = merged_data[(merged_data['smoke'] == 1) & (merged_data['alco'] == 1)].shape[0]
smoke_not_drink = merged_data[(merged_data['smoke'] == 1) & (merged_data['alco'] == 0)].shape[0]
drink_not_smoke = merged_data[(merged_data['smoke'] == 0) & (merged_data['alco'] == 1)].shape[0]
neither_smoke_nor_drink = merged_data[(merged_data['smoke'] == 0) & (merged_data['alco'] == 0)].shape[0]

# Create a DataFrame
data = {
    'Groups': ['Smoke and Drink', 'Smoke and Do Not Drink', 'Do Not Smoke and Drink', 'Do Not Smoke and Do Not Drink'],
    'Count': [smoke_and_drink, smoke_not_drink, drink_not_smoke, neither_smoke_nor_drink]
}
df = pd.DataFrame(data)

# Plotting bar chart
bars = plt.bar(df['Groups'], df['Count'], color=['purple', 'blue', 'red', 'green'])

# Customize the plot
plt.xlabel('Groups')
plt.ylabel('Amount')
plt.title('People Based on Smoking and Drinking Habits')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, int(yval), ha='center', va='bottom', color='black')

# Show the plot
plt.show()
```

```python
low_gluc = merged_data[(merged_data['gluc'] == 1) & (merged_data['cardio'] == 1)].shape[0]
norm_gluc = merged_data[(merged_data['gluc'] == 2) & (merged_data['cardio'] == 1)].shape[0]
high_gluc = merged_data[(merged_data['gluc'] == 3) & (merged_data['cardio'] == 1)].shape[0]

labels = ['Normal Glucose', 'High Glucose', 'Very High Glucose']
sizes = [low_gluc, norm_gluc, high_gluc]
colors = ['#ff9999','#66b3ff','#99ff99'] 

low_gluc2 = merged_data[(merged_data['gluc'] == 1) & (merged_data['cardio'] == 0)].shape[0]
norm_gluc2 = merged_data[(merged_data['gluc'] == 2) & (merged_data['cardio'] == 0)].shape[0]
high_gluc2 = merged_data[(merged_data['gluc'] == 3) & (merged_data['cardio'] == 0)].shape[0]

# Sample data for the second pie chart
labels2 = ['Normal Glucose', 'High Glucose', 'Very High Glucose']
sizes2 = [low_gluc2, norm_gluc2, high_gluc2]
colors2 = ['#ffb3e6', '#c2c2f0', '#ffb366']

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# First pie chart
axes[0].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
axes[0].set_title('Glucose of People With Cardiovascular Disease')

# Second pie chart
axes[1].pie(sizes2, labels=labels2, autopct='%1.1f%%', colors=colors2, startangle=90)
axes[1].set_title('Glucose of People Without Cardiovascular Disease')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
low_chol = merged_data[(merged_data['cholesterol'] == 1) & (merged_data['cardio'] == 1)].shape[0]
norm_chol = merged_data[(merged_data['cholesterol'] == 2) & (merged_data['cardio'] == 1)].shape[0]
high_chol = merged_data[(merged_data['cholesterol'] == 3) & (merged_data['cardio'] == 1)].shape[0]

labels = ['Normal Cholesterol', 'High Cholesterol', 'Very High Cholesterol']
sizes = [low_chol, norm_chol, high_chol]
colors = ['#ff9999','#66b3ff','#99ff99'] 

low_chol2 = merged_data[(merged_data['cholesterol'] == 1) & (merged_data['cardio'] == 0)].shape[0]
norm_chol2 = merged_data[(merged_data['cholesterol'] == 2) & (merged_data['cardio'] == 0)].shape[0]
high_chol2 = merged_data[(merged_data['cholesterol'] == 3) & (merged_data['cardio'] == 0)].shape[0]

# Sample data for the second pie chart
labels2 = ['Normal Cholesterol', 'High Cholesterol', 'Very High Cholesterol']
sizes2 = [low_chol2, norm_chol2, high_chol2]
colors2 = ['#ffb3e6', '#c2c2f0', '#ffb366']

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# First pie chart
axes[0].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
axes[0].set_title('Cholesterol of People With Cardiovascular Disease')

# Second pie chart
axes[1].pie(sizes2, labels=labels2, autopct='%1.1f%%', colors=colors2, startangle=90)
axes[1].set_title('Cholesterol of People Without Cardiovascular Disease')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your gender column has values 1 for male and 2 for female
gender_mapping = {1: 'Female', 2: 'Male'}

plt.figure(figsize=(12, 6))

# Plot the bar chart for ap_hi
plt.subplot(1, 2, 1)
sns.barplot(x='gender', y='ap_hi', data=merged_data.replace({'gender': gender_mapping}), color='blue', errorbar=None)
plt.title('Average ap_hi by Gender')
plt.ylim(80, 140)

# Manually set x-axis ticks to avoid swapping
plt.xticks(ticks=[0, 1], labels=list(gender_mapping.values()))

# Plot the bar chart for ap_lo
plt.subplot(1, 2, 2)
sns.barplot(x='gender', y='ap_lo', data=merged_data.replace({'gender': gender_mapping}), color='red', errorbar=None)
plt.title('Average ap_lo by Gender')
plt.ylim(75, 100)

# Manually set x-axis ticks to avoid swapping
plt.xticks(ticks=[0, 1], labels=list(gender_mapping.values()))

plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
gender_mapping = {1: 'Female', 2: 'Male'}

# Pie chart for Cardio distribution
plt.figure(figsize=(8, 8))
gender_counts = merged_data['gender'].map(gender_mapping).value_counts()
gender_counts.plot.pie(autopct='%1.2f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Prevalence of Cardiovascular Disease by Gender')
plt.legend(labels=gender_counts.index, loc='upper right')
plt.show()
```

```python

gender_mapping = {1: 'Male', 2: 'Female'}

# Grouped bar charts for health metrics by gender
health_metrics = ['ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'active', 'alco', 'cardio']

plt.figure(figsize=(16, 8))

for i, metric in enumerate(health_metrics, 1):
    plt.subplot(2, 4, i)
    ax = sns.barplot(x='gender', y=metric, data=merged_data.replace({'gender': gender_mapping}),
                     hue='gender', palette='pastel', err_kws={'linewidth': 0})
    
    plt.title(f'{metric.capitalize()} by Gender')
    
    # Add numbers on top of the bars with some adjustments
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=8, color='black')


plt.tight_layout()
plt.show()

```
